1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution
of age with respect to each gender along with the information about whether they survived or
not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above statistics.

```python
#import all necessary library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load inbuilt titanic data set
titanic_data=sns.load_dataset('titanic')
titanic_data
```

```python
#by using box plot without hue parameter
plt.figure(figsize=(5,5))
sns.boxplot(x='age',y='sex',data=titanic_data)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='survived', loc='upper right')
plt.show()
```

```python
#by using box plot with hue parameter

plt.figure(figsize=(6,6))
sns.boxplot(x='age',y='sex',hue='survived',data=titanic_data)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='survived',loc='upper right')
plt.show()
```
