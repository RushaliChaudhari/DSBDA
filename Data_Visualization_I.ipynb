```python
#import the library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the inbuilt titanic dataset
titanic_data=sns.load_dataset('titanic')
titanic_data
```

```python
sns.histplot(titanic_data['fare'], bins=20, kde=True, color='skyblue')
plt.xlabel('Ticket Fare')
plt.ylabel('Frequency')
plt.title('Distribution of Ticket Prices on Titanic')
plt.show()

```

```python
#draw boxplot
sns.boxplot(x='fare' , data=titanic_data,color='skyblue')
plt.xlabel('Ticket Fare')

plt.title('Distribution of Ticket Prices on Titanic')
plt.show()
```

```python
#scatter plot
sns.scatterplot(x=titanic_data['fare'], y=titanic_data['fare'], alpha=1)
plt.xlabel('Passenger Index')
plt.ylabel('Frequency')
plt.title('Distribution of Ticket Prices on Titanic')
plt.show()
```

```python
sns.countplot(x='fare',data=titanic_data)
plt.show()
```

```python
num_bins = 10

# Create bins for ticket fares
titanic_data['fare_bins'] = pd.cut(titanic_data['fare'], bins=num_bins)

# Plot count plot of ticket fare bins
plt.figure(figsize=(10, 6))
sns.countplot(x='fare_bins', data=titanic_data, palette='viridis')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel('Ticket Fare Bins')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Ticket Prices on Titanic')

# Show the plot
plt.tight_layout()
plt.show()

```
